## ДЗ №2

1. Вывести результат работы алгоритма PCA для датасета digits.

```python
from sklearn.datasets import load_digits

digits_dataset = load_digits()

digits = digits_dataset.data
digits
```

```python
digits_target = digits_dataset.target
digits_target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)

digits_pca = pca.fit_transform(digits)
digits_pca
```

```python
plt.scatter(digits_pca[:, 0], digits_pca[:, 1], c=digits_target)
```

2. Вывести результат работы алгоритма t-SNE для датасета digits.

```python
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2)

digits_tsne = tsne.fit_transform(digits)
digits_tsne
```

```python
plt.scatter(digits_tsne[:, 0], digits_tsne[:, 1], c=digits_target)
```

## ДЗ №3 Суперпиксельная сегментация изображения

```python
import matplotlib.pyplot as plt
from skimage.io import imread

image = imread('http://prorastet.ru/wp-content/uploads/2019/07/yutyuy.jpg')

plt.figure(figsize=(10, 10))
plt.imshow(image)
```

```python
import numpy as np
from sklearn.cluster import KMeans


def superpixel_segmentation(image: np.ndarray, cluster_count: int, k: float):
    new_image = np.empty((image.shape[0], image.shape[1], image.shape[2] + 2))

    for row in range(image.shape[0]):
        for column in range(image.shape[1]):
            new_image[row, column] = np.append(image[row, column], [row * k, column * k])

    image_flattened = new_image.reshape((-1, new_image.shape[-1]))
    kmeans = KMeans(n_clusters=cluster_count).fit(image_flattened)
    labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    cluster_colors = centroids[:, :-2]

    return cluster_colors[labels].astype(int).reshape(image.shape)
```

```python
k_list = [0, 0.2, 0.5, 0.8, 1, 1.5]
rows = 2
columns = 3

fig = plt.figure(figsize=(30, 30))

for i in range(len(k_list)):
    fig.add_subplot(rows, columns, i + 1)
    plt.imshow(superpixel_segmentation(image, 55, k_list[i]))
    plt.axis('off')
    plt.title(f"k = {k_list[i]}")
```
